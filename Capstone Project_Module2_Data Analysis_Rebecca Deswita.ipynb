##### Hasil Pencilan

1. Pencilan Semua Kolom Numerik


Pemeriksaan nilai pencilan pada dataset dilakukan pada kolom numerik, yaitu kolom dengan tipe data float64 dan int64. Ini dilakukan agar proses deteksi pencilan hanya diterapkan pada data yang relevan secara numerik.


Selanjutnya, didefinisikan sebuah fungsi bernama detect_outliers_iqr yang menggunakan metode Interquartile Range (IQR) untuk mendeteksi pencilan. Fungsi ini menghitung kuartil pertama (Q1) dan kuartil ketiga (Q3) seperti yang tercantum pada tabel statistik deskriptif di atas, lalu menentukan batas bawah dan atas dengan rumus:
- Batas Bawah = Q1 - 1.5 Ã— IQR
- Batas Atas = Q3 + 1.5 Ã— IQR

Data yang berada di luar batas tersebut dianggap sebagai pencilan. Fungsi ini mengembalikan jumlah pencilan serta nilai batas bawah dan atas untuk kolom yang dianalisa.


Kemudian, dilakukan fungsi looping pada semua kolom numerik di dataset. Untuk setiap kolom, fungsi detect_outliers_iqr dipanggil untuk menghitung pencilan, dan hasilnya disimpan dalam list "hasil_pencilan" sebagai dictionary berisi nama kolom, jumlah pencilan, dan nilai batas bawah dan atas.

```python
# Ambil hanya kolom numerik
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns

# Fungsi deteksi outlier IQR
def detect_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    batas_bawah = Q1 - 1.5 * IQR
    batas_atas = Q3 + 1.5 * IQR
    pencilan = data[(data[column] < batas_bawah) | (data[column] > batas_atas)]
    return {
        'Jumlah Pencilan': len(pencilan),
        'Batas Bawah': batas_bawah,
        'Batas Atas': batas_atas
    }

# Loop ke semua kolom numerik dan simpan hasilnya
hasil_pencilan = []

for col in numeric_cols:
    hasil = detect_outliers_iqr(df, col)
    hasil_pencilan.append({
        'Kolom': col,
        'Jumlah Pencilan': hasil['Jumlah Pencilan'],
        'Batas Bawah': hasil['Batas Bawah'],
        'Batas Atas': hasil['Batas Atas']
    })

# Tampilkan hasil
df_pencilan = pd.DataFrame(hasil_pencilan)
print(df_pencilan)
```

2. Pencilan Kolom Relevan


Dilakukan pemeriksaan kembali terhadap nilai pencilan pada kolom yang dianggap relevan terhadap tujuan proyek ini, yakni kolom 'trip_distance', 'fare_amount', 'tip_amount', 'tolls_amount', 'total_amount'. Formula kode yang digunakan sama dengan bagian 1 di atas saat mendeteksi pencilan pada tiap kolom numerik.

```python
# Daftar kolom yang ingin dianalisa untuk pencilan
kolom_terpilih = ['trip_distance', 'fare_amount', 'tip_amount', 'tolls_amount', 'total_amount']

# Fungsi deteksi pencilan menggunakan metode IQR
def ambil_pencilan_iqr(data, kolom):
    Q1 = data[kolom].quantile(0.25)
    Q3 = data[kolom].quantile(0.75)
    IQR = Q3 - Q1
    batas_bawah = Q1 - 1.5 * IQR
    batas_atas = Q3 + 1.5 * IQR
    pencilan = data[(data[kolom] < batas_bawah) | (data[kolom] > batas_atas)]
    return pencilan, batas_bawah, batas_atas

# Siapkan tabel hasil deteksi pencilan per kolom
hasil_pencilan_terpilih = []

# Loop ke kolom terpilih dan simpan hasilnya dalam tabel ringkasan
for kolom in kolom_terpilih:
    if kolom in df.columns:
        pencilan, batas_bawah, batas_atas = ambil_pencilan_iqr(df, kolom)
        hasil_pencilan_terpilih.append({
            'Kolom': kolom,
            'Jumlah Pencilan': len(pencilan),
            'Batas Bawah': batas_bawah,
            'Batas Atas': batas_atas
        })

# Buat DataFrame hasil
tabel_pencilan_terpilih = pd.DataFrame(hasil_pencilan_terpilih)

# Tampilkan hasil
print(tabel_pencilan_terpilih)
```

A. Efisiensi Pendapatan 

Untuk mendukung eksplorasi data terkait performa pendapatan dan efisiensi operasional supir taksi, beberapa fitur turunan dihitung dan ditambahkan ke dataset df_cleaned. Fitur-fitur ini berfungsi untuk mengevaluasi berapa efektif supir dalam menghasilkan uang berdasarkan durasi, jarak, dan komponen pendapatan tambahan seperti tip, serta membedakan pola performa antara hari kerja dan akhir pekan.

- trip_duration_min: dihitung dari selisih waktu antara lpep_dropoff_datetime dan lpep_pickup_datetime, diubah ke satuan menit. Kolom ini digunakan untuk mengetahui durasi tiap perjalanan sebagai dasar analisa waktu.

- fare_per_minute: dihitung dengan membagi fare_amount dengan trip_duration_min. Fitur ini mengukur seberapa efisien waktu kerja supir dalam menghasilkan pendapatan, atau bisa dikatakan sebagai pendapatan per menit.

- fare_per_mile: dihitung dengan membagi fare_amount dengan trip_distance. Kolom ini digunakan untuk menilai efisiensi jarak, yaitu berapa besar pendapatan yang dihasilkan per mil perjalanan.

- revenue_per_trip: merupakan penjumlahan dari fare_amount dan tip_amount. Kolom ini mewakili total pendapatan aktual supir per perjalanan dari sisi penumpang (tidak termasuk biaya tambahan seperti tol atau surcharge).

- is_weekend: kolom logika (True/False) yang menunjukkan apakah perjalanan terjadi di akhir pekan (Sabtu atau Minggu). Fitur ini ditambahkan untuk memungkinkan analisa performa supir berdasarkan hari, terutama dalam membandingkan pendapatan antara hari kerja dan akhir pekan.

```python
# trip_duration_min (Durasi perjalanan dalam menit)
df_cleaned['trip_duration_min'] = (df_cleaned['lpep_dropoff_datetime'] - df_cleaned['lpep_pickup_datetime']).dt.total_seconds() / 60

# fare_per_minute (Efisiensi waktu â†’ berapa dolar per menit)
df_cleaned['fare_per_minute'] = df_cleaned['fare_amount'] / df_cleaned['trip_duration_min']

# fare_per_mile (Efisiensi jarak â†’ berapa dolar per mil)
df_cleaned['fare_per_mile'] = df_cleaned['fare_amount'] / df_cleaned['trip_distance']

# revenue_per_trip (Pendapatan supir per trip dari fare + tip)
df_cleaned['revenue_per_trip'] = df_cleaned['fare_amount'] + df_cleaned['tip_amount']
```

```python
# pickup_date (Tanggal penjemputan dari waktu pickup, untuk agregasi harian)
df_cleaned['pickup_date'] = df_cleaned['lpep_pickup_datetime'].dt.date

# pickup_hour (Jam penjemputan dalam format 0â€“23, untuk analisis jam sibuk)
df_cleaned['pickup_hour'] = df_cleaned['lpep_pickup_datetime'].dt.hour

# pickup_day_name (Nama hari penjemputan seperti 'Monday', 'Tuesday', dst.)
df_cleaned['pickup_day_name'] = df_cleaned['lpep_pickup_datetime'].dt.day_name()

# dropoff_date (Tanggal pengantaran, berguna untuk analisis perjalanan lintas hari)
df_cleaned['dropoff_date'] = df_cleaned['lpep_dropoff_datetime'].dt.date

# dropoff_hour (Jam pengantaran dalam format 0â€“23, untuk melihat jam selesai trip)
df_cleaned['dropoff_hour'] = df_cleaned['lpep_dropoff_datetime'].dt.hour

# dropoff_day_name (Nama hari pengantaran seperti 'Monday', 'Tuesday', dst.)
df_cleaned['dropoff_day_name'] = df_cleaned['lpep_dropoff_datetime'].dt.day_name()

# is_weekend (untuk analisa hari kerja vs akhir pekan)
df_cleaned['is_weekend'] = df_cleaned['pickup_day_name'].isin(['Saturday', 'Sunday'])
```

##### b. Statistik Inferensial

```python
from scipy.stats import ttest_ind

# Buat kolom weekend vs weekday jika belum ada
df_cleaned['is_weekend'] = df_cleaned['pickup_day'].isin(['Saturday', 'Sunday'])

# Pisahkan data menjadi dua grup
fare_weekend = df_cleaned[df_cleaned['is_weekend']]['fare_per_minute']
fare_weekday = df_cleaned[~df_cleaned['is_weekend']]['fare_per_minute']

# Lakukan uji t (equal_var=False untuk Welchâ€™s t-test)
t_stat, p_value = ttest_ind(fare_weekend, fare_weekday, equal_var=False)

# Tampilkan hasil
print(f"Hasil Uji T (Pendapatan per Menit - Akhir Pekan vs Hari Biasa):")
print(f"T-Statistic: {t_stat:.2f}")
print(f"P-Value: {p_value:.2e}")  
```

Intepretasi Hasil Uji T (T-test):
- Karena p-value < 0.05, maka hasilnya signifikan secara statistik.

- Artinya pendapatan per menit pada akhir pekan benar-benar lebih tinggi secara nyata dibandingkan hari biasa.

- Pengemudi taksi sebaiknya mengutamakan beroperasi pada akhir pekan untuk efisiensi waktu kerja maksimal.

```python
from scipy.stats import f_oneway

# Pastikan pickup_day sudah dalam format yang benar
df_cleaned['pickup_day'] = df_cleaned['lpep_pickup_datetime'].dt.day_name()

# Kelompokkan data fare berdasarkan nama hari
fare_by_day = [group['fare_amount'].values for _, group in df_cleaned.groupby('pickup_day')]

# Lakukan One-Way ANOVA
f_stat, p_value = f_oneway(*fare_by_day)

# Tampilkan hasil
print("Hasil Uji ANOVA (Pendapatan per Perjalanan antar Hari):")
print(f"F-Statistic: {f_stat:.2f}")
print(f"P-Value: {p_value:.4f}")
```

Interpretasi Hasil Uji Anova:
- Terdapat perbedaan signifikan secara statistik dalam rata-rata pendapatan per perjalanan antara hari-hari dalam seminggu.

- Rata-rata pendapatan pengemudi taksi per perjalanan tidak sama sepanjang minggu.

- Hari-hari tertentu, terutama Sabtu dan Minggu, memiliki rata-rata pendapatan per perjalanan secara konsisten lebih tinggi dibanding hari biasa (Seninâ€“Jumat).
